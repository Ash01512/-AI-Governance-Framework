Data preprocessing¶


```python
# Transform trending_date to pandas datetime format
us_videos['trending_date'] = pd.to_datetime(us_videos['trending_date'], format='%y.%d.%m')
us_videos.trending_date.value_counts().sort_index(inplace=True)
us_videos.head()
```

```python
# Check if 'publish_time' column exists before trying to process it
if 'publish_time' in us_videos.columns:
    # Transforming publish_time to datetime and storing as publish_date
    us_videos['publish_date'] = pd.to_datetime(us_videos['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
    # Drop original publish_time column as it's no longer needed
    us_videos.drop('publish_time', axis=1, inplace=True)
elif 'publish_date' in us_videos.columns:
    # If publish_time is already dropped, ensure publish_date is of datetime type
    us_videos['publish_date'] = pd.to_datetime(us_videos['publish_date'])
```

```python
# Create New Variable Counting Days to Achieving Trending Status
us_videos['days_to_trending'] = (pd.to_datetime(us_videos.trending_date) - pd.to_datetime(us_videos.publish_date)).dt.days
us_videos.days_to_trending.describe(percentiles=[.05,.25,.5,.75,.95])
```

Create Meaningful Data Frame Index¶


```python
us_videos.set_index(['trending_date','video_id'],inplace=True)
us_videos.head()
```

Dislike percentage¶


```python
us_videos['dislike_percentage'] = us_videos['dislikes'] / (us_videos['dislikes'] + us_videos['likes'])
us_videos.dislike_percentage.describe(percentiles=[.05,.25,.5,.75,.95])
```

Data Clean Up¶
Drop Unreliable Observations

```python
# First, convert 'video_error_or_removed' to boolean, filling NaN values with False
us_videos['video_error_or_removed'] = us_videos['video_error_or_removed'].fillna(False).astype(bool)

# Now, interpret 'video_error_or_removed' == True
print(us_videos[us_videos.video_error_or_removed])
us_videos = us_videos[~us_videos.video_error_or_removed]
```

```python
# Create 'publish_date' column from 'publish_time' and ensure it's in datetime format.
us_videos['publish_date'] = pd.to_datetime(us_videos['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```
